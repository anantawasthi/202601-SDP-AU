## **Introduction to Git**

Git is a distributed version control system (DVCS) designed to handle everything from small to very large projects with speed and efficiency. It allows multiple developers to work on a project simultaneously without overwriting each other's changes. Git tracks changes to files, enabling users to revert to previous versions, branch off to develop features independently, and merge changes back into the main codebase.

## **History of Git**

Git was created by Linus Torvalds in 2005 for the development of the Linux kernel. Before Git, the Linux kernel community used a proprietary DVCS called BitKeeper. When the free-of-charge status for BitKeeper was revoked, Torvalds set out to create his own version control system that was distributed, fast, and secure. He developed Git in just a few days, and it quickly became the standard for the Linux kernel and many other open-source projects.

## **Introduction to GitHub**

GitHub is a web-based platform that uses Git for version control. It provides hosting for software development and version control using Git, offering all of the distributed version control and source code management (SCM) functionality of Git as well as adding its own features. These include access control, bug tracking, feature requests, task management, continuous integration, and wikis for every project. GitHub is widely used for collaborative software development and is a popular platform for hosting open-source projects.

## **GitHub in Action**

## **Git Clone**

The `git clone` command is used to create a local copy of a remote Git repository. When you clone a repository, you download all the project files, versions, and branches to your local machine.

**Syntax:**

```bash
git clone <repository_url>
```

**Example:**

To clone a repository from GitHub:

```bash
git clone https://github.com/anantawasthi/202601-SDP-AU.git
```

This command will create a new directory named `202601-SDP-AU` in your current working directory, containing a full copy of the remote repository.

## **Git Add**

The `git add` command is used to add changes in the working directory to the staging area. The staging area is a place where you prepare changes before committing them. It allows you to select which changes (from modified or new files) you want to include in the next commit.

**Syntax:**

```bash
git add <file_name>
```
or to add all changes in the current directory:

```bash
git add .
```

**Examples:**

1. To add a specific file:

   ```bash
   git add index.html
   ```

2. To add multiple specific files:

   ```bash
   git add style.css script.js
   ```

3. To add all changes (new files and modifications) in the current directory and its subdirectories:

   ```bash
   git add .
   ```

4. To add all changes (new files and modifications) in the current directory (excluding deleted files):

   ```bash
   git add -A
   ```

## **Git Commit**

The `git commit` command is used to save the changes from the staging area to the local repository. Each commit represents a snapshot of your project at a specific point in time and includes a unique identifier (hash) and a commit message.

**Syntax:**

```bash
git commit -m "<commit_message>"
```

**Example:**

After staging your changes with `git add`:

```bash
git commit -m "Add initial homepage structure and styling"
```

This command will create a new commit in your local repository with the provided message. The commit message should be descriptive, explaining what changes were made in that commit.

## **Git Push**

The `git push` command is used to upload your local commits to a remote repository (e.g., GitHub, GitLab). This makes your changes available to others and synchronizes your local repository with the remote one.

**Syntax:**

```bash
git push <remote_name> <branch_name>
```

Often, for the first push of a new branch, you might use the `-u` flag to set the upstream tracking branch:

```bash
git push -u <remote_name> <branch_name>
```

**Examples:**

1. To push changes from your local `main` branch to the `origin` remote repository:

   ```bash
   git push origin main
   ```

2. To set the upstream for a new branch named `feature-x` and push it to `origin`:

   ```bash
   git push -u origin feature-x
   ```

   After setting the upstream, subsequent pushes on that branch can be simplified:

   ```bash
   git push
   ```

## **Git Branches**

In Git, a branch is essentially a lightweight, movable pointer to a commit. The default branch name is usually `main` (or `master` in older repositories). Branches allow you to diverge from the main line of development and continue to work independently without affecting the main project.

**Why use branches?**

*   **Parallel Development:** Multiple developers can work on different features or bug fixes simultaneously without interfering with each other's code.
*   **Experimentation:** You can experiment with new ideas or features in a separate branch without risking the stability of the main codebase.
*   **Isolation:** Changes made in one branch are isolated from others until they are explicitly merged.

When you create a new branch, it essentially points to the same commit as the branch you were on. As you make new commits on your new branch, the branch pointer moves forward. Later, you can merge these changes back into your main branch or another branch.

## **Git Checkout**

The `git checkout` command is primarily used to switch between different branches or to restore working tree files. It allows you to navigate through the project's history and work on different lines of development.

**Syntax:**

1. **To switch to an existing branch:**

   ```bash
   git checkout <branch_name>
   ```

2. **To create and switch to a new branch:**

   ```bash
   git checkout -b <new_branch_name>
   ```

3. **To restore a file to its state from a specific commit or the staging area:**

   ```bash
   git checkout <commit_hash> <file_path>
   ```
   or
   ```bash
   git checkout -- <file_path>
   ```

**Examples:**

1. To switch to a branch named `feature-x`:

   ```bash
   git checkout feature-x
   ```

2. To create a new branch named `new-feature` and switch to it:

   ```bash
   git checkout -b new-feature
   ```

3. To discard changes in `index.html` and revert it to its last committed state (or staged state):

   ```bash
   git checkout -- index.html
   ```

## **Git Merge**

The `git merge` command is used to integrate changes from one branch into another. It combines the commit histories of two or more branches into a single new commit, effectively bringing the changes from one branch into the target branch.

**Syntax:**

```bash
git merge <source_branch>
```

**Example:**

Let's say you are on your `main` branch and you want to bring in the changes from a `feature-x` branch:

1. First, ensure you are on the target branch (e.g., `main`):

   ```bash
   git checkout main
   ```

2. Then, merge the `feature-x` branch into `main`:

   ```bash
   git merge feature-x
   ```

If there are no conflicts, Git will automatically create a new merge commit. If conflicts occur (when both branches modified the same part of the same file), you will need to resolve them manually before completing the merge with `git commit`.
