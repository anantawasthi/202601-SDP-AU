
**1) Proprietary Frontier Models** – top AI models focused on maximal performance, often behind managed APIs or products

* **OpenAI** – GPT-5 and the *o4* family (successors to GPT-4) — flagship general/intelligent LLMs used in ChatGPT and APIs.
* **Anthropic** – Claude series (e.g., Claude 4 Opus, Claude 4 Sonnet) — strong safety and reasoning-oriented models.
* **Google DeepMind/Google** – *Gemini* family (e.g., Gemini 2.5 Pro, Flash) — multimodal and broad general AI.
* **xAI (Elon Musk)** – *Grok* (e.g., Grok 3/4) — conversationally oriented, used in social/real-time settings.
* **Alibaba** – *Qwen* (latest 3 family) — large commercial models popular in Asia.

**2) Open-Source / Open-Weight Models** – freely usable and self-hostable models with wide community adoption

* **Meta (Meta AI)** – *LLaMA* family (e.g., Llama 4 Scout/Maverick) — partially open source, flexible for research and fine-tuning.
* **DeepSeek** – DeepSeek V and R series (e.g., DeepSeek-V3.2 / R1) — open-source models competitive with commercial offerings.
* **OpenAI (Open-weight)** – *gpt-oss-120B* series — open-weight variants suitable for localized deployment.
* **GLM (ZAI)** – GLM-4.6 — high-context, general-purpose open-source LLM.

**3) Specialized / Domain or Research Models** – models tailored for specific tasks or with academic transparency

* **Various contributors (community/academic)** – *Falcon* (ATRC/UAE), *Baichuan 2* (Alibaba research open), *TigerBot*, *YAYI*, etc. — focus on multilingual, research use cases, or niche domains.

**Broad Company Overview (LLM Leaders)**

* **OpenAI** (USA) — GPT family, commercial leader.

* **Google/DeepMind** (USA) — Gemini models with multimodal strengths.
