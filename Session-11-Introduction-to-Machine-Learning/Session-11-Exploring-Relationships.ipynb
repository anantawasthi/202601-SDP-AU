## Summary of Statistical Methods for Dependent (Y) and Independent (X) Variables by Measurement Scale

This table provides a quick reference for selecting appropriate statistical methods to establish relationships between a dependent variable (Y) and an independent variable (X), considering their respective measurement scales.

| Dependent Variable (Y) Scale | Independent Variable (X) Scale | Recommended Statistical Methods                                                                                                                                                                                                                                                                                                                                                                                                                                                                 |
| :--------------------------- | :----------------------------- | :-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Nominal**                  | **Nominal**                    | Chi-square test of independence, Fisher's exact test (for small samples), Measures of Association (Cramer's V, Phi coefficient, Lambda).                                                                                                                                                                                                                                                                                                                                                      |
| **Nominal**                  | **Ordinal**                    | Chi-square test of independence (treating ordinal as categories), Non-parametric tests (Mann-Whitney U test for binary Y and few ordinal X categories, Kruskal-Wallis H test for multiple nominal Y categories and ordinal X), Logistic Regression (treating ordinal as a predictor).                                                                                                                                                                                                        |
| **Nominal**                  | **Scale**                      | Logistic Regression (Binary, Multinomial, or Ordinal Logistic Regression depending on Y's nature), Discriminant Analysis (Linear or Quadratic).                                                                                                                                                                                                                                                                                                                                                            |
| **Ordinal**                  | **Nominal**                    | Non-parametric tests (Mann-Whitney U test for two nominal groups, Kruskal-Wallis H test for more than two nominal groups), Ordinal Logistic Regression.                                                                                                                                                                                                                                                                                                                                                         |
| **Ordinal**                  | **Ordinal**                    | Spearman's Rank-Order Correlation, Kendall's Tau, Ordinal Logistic Regression.                                                                                                                                                                                                                                                                                                                                                                                                                                |
| **Ordinal**                  | **Scale**                      | Ordinal Logistic Regression. Less recommended: Categorizing the Scale X and then using methods for ordinal-ordinal relationships (with caution due to information loss).                                                                                                                                                                                                                                                                                                                      |
| **Scale**                    | **Nominal**                    | t-test (for two nominal groups), ANOVA (Analysis of Variance) for three or more nominal groups. Non-parametric alternatives: Mann-Whitney U test (for two nominal groups), Kruskal-Wallis H test (for three or more nominal groups) if parametric assumptions are violated.                                                                                                                                                                                                                    |
| **Scale**                    | **Ordinal**                    | Spearman's Rank-Order Correlation (if X is treated as ranked), Linear Regression (treating ordinal categories as ordered numerical values, or using dummy variables), Non-parametric ANOVA (Kruskal-Wallis H test) if parametric assumptions are violated and X defines distinct groups.                                                                                                                                                                                                             |
| **Scale**                    | **Scale**                      | Pearson Correlation (for linear relationships), Linear Regression (for predicting Y from X). Non-parametric alternatives: Spearman's Rank-Order Correlation (if assumptions for Pearson's are not met or for monotonic but non-linear relationships), Advanced Regression Techniques (for non-linear relationships if applicable). |
